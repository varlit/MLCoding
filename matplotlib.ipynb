```python
import  matplotlib.pyplot as plt
plt.figure(figsize=(10,10), dpi=100)
plt.figure(1)
colorType = ['or','+b']
i = 0
for a in range(0,2):
    pa = colorType[a]
    #print(pa)
    for b in range(0,2):
        pb = colorType[b]
        for c in range(0,2):
            pc = colorType[c]
            for d in range(0,2):
                pd = colorType[d]
                i = i + 1
                # print(i)
                ax = plt.subplot(4,4,i)
                plt.plot(0,0,pa)
                plt.plot(0,1,pb)
                plt.plot(1,0,pc)
                plt.plot(1,1,pd)                
```
